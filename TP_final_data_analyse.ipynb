```python
#On crée une copie de la table population
df_population_1 = df_population
#On supprime les données déclarées en double
df_population_1 = df_population_1.drop(33, axis=0).reset_index()
#on garde que les champs utiles
df_population_1 = df_population_1[['Zone', 'Code zone', 'Valeur']]
df_population_1

```

```python
#On supprime les doublons et on calcul la population mondiale en 2013
df_population_1 = df_population_1.drop_duplicates()
population_mondiale = str(df_population_1['Valeur'].sum(skipna=True)* 1000)
print("Population totale en 2013 : " + population_mondiale)
```

```python
df_animaux_1 = df_animaux
df_animaux_1['origine'] = 'animal'
df_vegetaux_1 = df_veg
df_vegetaux_1['origine'] = 'vegetal'
df_ani_veg = df_animaux_1.append(df_vegetaux_1)

#maskhatch biha 
#idx = np.where(( (df_ani_veg['Élément']=="Production") | (df_ani_veg['Produit']=="Viande de Bovins") ))
# 'Disponibilité alimentaire (Kcal/personne/jour)',
 # 'Disponibilité de protéines en quantité (g/personne/jour)']), 

df_ani_veg_clean1 = df_ani_veg.loc[df_ani_veg['Élément']=='Disponibilité alimentaire en quantité (kg/personne/an)']
df_ani_veg_clean2 = df_ani_veg.loc[df_ani_veg['Élément']=='Disponibilité alimentaire (Kcal/personne/jour)']
df_ani_veg_clean3 = df_ani_veg.loc[df_ani_veg['Élément']=='Disponibilité de protéines en quantité (g/personne/jour)']
df_ani_veg_clean4 = df_ani_veg.loc[df_ani_veg['Élément']=='Disponibilité intérieure']
df_ani_veg_clean5 = df_ani_veg.loc[df_ani_veg['Élément']=='Nourriture']

df_ani_veg_clean6 = df_ani_veg_clean1.append(df_ani_veg_clean2)
df_ani_veg_clean7 = df_ani_veg_clean6.append(df_ani_veg_clean3)
df_ani_veg_clean8 = df_ani_veg_clean7.append(df_ani_veg_clean4)
df_ani_veg_clean = df_ani_veg_clean8.append(df_ani_veg_clean5)

df_ani_veg_clean = pd.pivot_table(df_ani_veg_clean, index=['Code zone', 'Zone', 'Code Produit', 'Produit', 'origine'], columns = 'Élément', values = 'Valeur', aggfunc=np.sum, fill_value=0).reset_index()

df_ani_veg_clean = pd.merge(df_ani_veg_clean, df_population_1, on =['Zone', 'Code zone'])

df_ani_veg_clean['dispo_alimentaire_kg']=df_ani_veg_clean['Disponibilité alimentaire en quantité (kg/personne/an)']*df_ani_veg_clean['Valeur']*1000
df_ani_veg_clean['dispo_alimentaire_kcal']=df_ani_veg_clean['Disponibilité alimentaire (Kcal/personne/jour)']*df_ani_veg_clean['Valeur']*1000*365
df_ani_veg_clean['dispo_protéine_Kg']=df_ani_veg_clean['Disponibilité de protéines en quantité (g/personne/jour)']*df_ani_veg_clean['Valeur']*365
#df_ani_veg_clean = df_ani_veg_clean.drop_duplicates().reset_index()
df_ani_veg_clean.head()
```

```python
df_ani_veg_clean['Ratio énergie/poids']=df_ani_veg_clean['dispo_alimentaire_kcal']/df_ani_veg_clean['dispo_alimentaire_kg']
df_ani_veg_clean.head()
```

```python
df_ani_veg_clean.groupby('Produit',as_index=False).agg({"Ratio énergie/poids": "mean"}).sort_values(by="Ratio énergie/poids",ascending=False).replace([np.inf, -np.inf], np.nan).dropna().reset_index()[0:20]
```

```python
df_ani_veg_clean['Protéine%']=(df_ani_veg_clean['dispo_protéine_Kg']/df_ani_veg_clean['dispo_alimentaire_kg'])*100

df_ani_veg_clean.groupby('Produit',as_index=False).agg({"Protéine%": "mean"}).sort_values(by="Protéine%",ascending=False).replace([np.inf, -np.inf], np.nan).dropna().reset_index()[0:20]
```

```python
#Questio 7
df1=df_ani_veg_clean[df_ani_veg_clean['origine']=='vegetal']
df1 = df1.replace([-np.inf, np.inf], np.nan)
df1
```

```python
df1['Dispo_int_kcal']=df1['Disponibilité intérieure']*1000000* df1['Ratio énergie/poids']

df1
```

```python
#question 7 
Somme_dispo_interne=round(df1['Dispo_int_kcal'].sum(skipna=True),2)
print('Disponibilité intérieure issue des produits d\'origine végétale : ' + str(Somme_dispo_interne) + ' kcal.')

```

```python
#question 8
# On considère que la consommation moyenne en calories d'un être humain est de 2000 kcal
nombre_humains_kcal=df1['Dispo_int_kcal'].sum(skipna=True)/ (2000*365)
pourcentage_humains_kcal=nombre_humains_kcal/(df_population_1['Valeur'].sum(skipna=True)* 1000)

print('Pourcentage d\'humains nourris avec les calories de la disponibilité intérieure: ' 
     + str(pourcentage_humains_kcal*100) + " %")
```

```python
df1['dispo_int_prot']=(df1['Disponibilité intérieure'] * 1000000) * (df1['Protéine%']/100)
dispo_int_veg_prot = round(df1['dispo_int_prot'].sum(skipna=True),2)
print('Disponibilité intérieure issue des produits d\'origine végétale : ' + str(dispo_int_veg_prot) + ' kg de protéines.')
```

```python
# On considère que la consommation moyenne en protéines d'un être humain est de 50 g par jour
nbre_humains_proteine = dispo_int_veg_prot / (50 * 1000 * 365)
pourcentage_humains_proteine = (nbre_humains_proteine / df_population_1['Valeur'].sum(skipna=True)* 1000)
print('Pourcentage d\'humains nourris  avec les protéines de la disponibilité intérieure: '  
     + str(pourcentage_humains_proteine) +  ' %')
```

```python
#question 9
df_ani_veg_clean['Nourriture_Kcal']=df_ani_veg_clean['Nourriture']*1000000*df_ani_veg_clean['Ratio énergie/poids']
df_ani_veg_clean = df_ani_veg_clean.replace([-np.inf, np.inf], np.nan)
# On somme sur la colonne de 'Nourriture_Kcal' pour avoir la nouriture disponible mondialement 
dispo_Nour_total_Kcal = df_ani_veg_clean['Nourriture_Kcal'].sum(skipna=True)
print('Nourriture totale disponible mondialement : ' + str(dispo_Nour_total_Kcal) + ' kcal.')

```

```python
# On considère que la consommation moyenne en calories d'un être humain est de 2000 kcal
nombre_humains_kcal=dispo_Nour_total_Kcal/ (2000*365)
pourcentage_humains_kcal=nombre_humains_kcal/(df_population_1['Valeur'].sum(skipna=True)* 1000)
print('Pourcentage d\'humains nourris avec les calories de la disponibilité alimentaire: ' + str(pourcentage_humains_kcal*100) + " %")
```

```python
# question 9 proteines
df_ani_veg_clean['Nourriture_prot']=df_ani_veg_clean['Nourriture']*1000000*df_ani_veg_clean['Protéine%']
# On somme sur la colonne de 'Nourriture_Kcal' pour avoir la nouriture disponible mondialement 
dispo_Nour_total_prot = df_ani_veg_clean['Nourriture_prot'].sum(skipna=True)
print('Nourriture totale disponible mondialement en proteines : ' + str(dispo_Nour_total_prot) + '.')

# On considère que la consommation moyenne en protéines d'un être humain est de 50 g par jour

nbre_humains_proteine = dispo_Nour_total_prot / (50 * 1000 * 365)
pourcentage_humains_proteine = (nbre_humains_proteine / df_population_1['Valeur'].sum(skipna=True)* 1000)
print('Pourcentage d\'humains nourris  avec les protéines de la disponibilité intérieure: '  
     + str(pourcentage_humains_proteine) +  ' %')
```
